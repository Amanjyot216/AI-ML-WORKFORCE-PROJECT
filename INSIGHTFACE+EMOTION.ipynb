```python
import cv2
import os
import numpy as np
from insightface.app import FaceAnalysis
from deepface import DeepFace
from collections import defaultdict

#
# CONFIG
#

VIDEO_PATH = "Video 4.mp4"

OUTPUT_DIR = "face_frames_out"
VIDEO_OUT_DIR = "member_videos_out"

os.makedirs(OUTPUT_DIR, exist_ok=True)
os.makedirs(VIDEO_OUT_DIR, exist_ok=True)

FRAME_SKIP = 10
PADDING = 20
OUTPUT_SIZE = 512

THRESHOLD = 0.40
VOTE_REQUIRED = 2

MIN_FACE_RATIO = 0.06

#
# MEMBERS
#

MEMBERS = [
    "Akshay", "Anjali", "Kasturi", "Om", "Bhima",
    "Amanjyot", "Sonali", "SSonali", "Shruti", "Sunny",
    "Aishwarya", "Hanmant", "Rohan", "Bhagyashree"
]

REFERENCE_IMAGES = {
    name: [f"/content/{name}{i}.png" for i in range(1, 6)]
    for name in MEMBERS
}

#
# FACE ENHANCEMENT PIPELINE
#

def enhance_face(img):
    img = cv2.resize(img, (OUTPUT_SIZE, OUTPUT_SIZE),
                     interpolation=cv2.INTER_LANCZOS4)

    img = cv2.bilateralFilter(img, d=5, sigmaColor=35, sigmaSpace=35)

    img = cv2.fastNlMeansDenoisingColored(
        img, None,
        h=4,
        hColor=4,
        templateWindowSize=7,
        searchWindowSize=21
    )

    lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
    l, a, b = cv2.split(lab)

    clahe = cv2.createCLAHE(clipLimit=1.0, tileGridSize=(8, 8))
    l = clahe.apply(l)

    img = cv2.cvtColor(cv2.merge([l, a, b]), cv2.COLOR_LAB2BGR)

    kernel = np.array([
        [0, -0.07, 0],
        [-0.07, 1.25, -0.07],
        [0, -0.07, 0]
    ])
    img = cv2.filter2D(img, -1, kernel)

    img = cv2.GaussianBlur(img, (3, 3), 0.3)

    return img


#
# EMOTION DETECTION
#

def detect_emotion(face_img):
    try:
        result = DeepFace.analyze(
            face_img,
            actions=['emotion'],
            enforce_detection=False,
            detector_backend='skip'
        )
        return result[0]['dominant_emotion']
    except:
        return "unknown"


#
# INIT INSIGHTFACE
#

print("\n[INFO] Loading InsightFace model...\n")

app = FaceAnalysis(name="buffalo_l")
app.prepare(ctx_id=0, det_size=(640, 640))


#
# LOAD REFERENCE EMBEDDINGS
#

print("[INFO] Extracting reference embeddings...\n")

reference_db = defaultdict(list)

for person, img_list in REFERENCE_IMAGES.items():

    for path in img_list:

        img = cv2.imread(path)
        if img is None:
            continue

        faces = app.get(img)
        if len(faces) == 0:
            continue

        reference_db[person].append(faces[0].embedding)

    print(f"{person}: {len(reference_db[person])} references loaded")

print("\n[INFO] Reference loading complete.\n")


#
# COSINE SIMILARITY
#

def cosine_sim(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))


#
# VIDEO WRITERS
#

video_writers = {}

def get_writer(person, fps=10):

    if person in video_writers:
        return video_writers[person]

    out_path = os.path.join(VIDEO_OUT_DIR, f"{person}.mp4")

    fourcc = cv2.VideoWriter_fourcc(*"mp4v")
    writer = cv2.VideoWriter(
        out_path,
        fourcc,
        fps,
        (OUTPUT_SIZE, OUTPUT_SIZE)
    )

    video_writers[person] = writer
    return writer


#
# MAIN VIDEO LOOP
#

cap = cv2.VideoCapture(VIDEO_PATH)

frame_id = 0
votes = defaultdict(int)
saved_count = defaultdict(int)

print("\n[INFO] Starting recognition...\n")

while True:

    ret, frame = cap.read()
    if not ret:
        break

    frame_id += 1
    if frame_id % FRAME_SKIP != 0:
        continue

    H, W, _ = frame.shape

    faces = app.get(frame)

    for face in faces:

        x1, y1, x2, y2 = face.bbox.astype(int)
        face_h = y2 - y1

        if face_h < MIN_FACE_RATIO * H:
            continue

        x1p = max(0, x1 - PADDING)
        y1p = max(0, y1 - PADDING)
        x2p = min(W, x2 + PADDING)
        y2p = min(H, y2 + PADDING)

        crop = frame[y1p:y2p, x1p:x2p]
        emb = face.embedding

        best_person = None
        best_score = -1

        for person, refs in reference_db.items():
            if len(refs) == 0:
                continue

            score = max(cosine_sim(emb, r) for r in refs)

            if score > best_score:
                best_score = score
                best_person = person

        if best_score < THRESHOLD:
            continue

        votes[best_person] += 1

        if votes[best_person] >= VOTE_REQUIRED:

            enhanced = enhance_face(crop)
            emotion = detect_emotion(enhanced)

            display_img = enhanced.copy()

            # Red border
            cv2.rectangle(display_img, (0, 0),
                          (OUTPUT_SIZE - 1, OUTPUT_SIZE - 1),
                          (0, 0, 255), 3)

            # Name + similarity (top-left)
            label_left = f"{best_person} | {best_score:.2f}"
            cv2.putText(display_img, label_left,
                        (10, 25),
                        cv2.FONT_HERSHEY_SIMPLEX,
                        0.7, (0, 0, 255), 2)

            # Emotion (top-right)
            label_right = emotion
            text_size = cv2.getTextSize(
                label_right,
                cv2.FONT_HERSHEY_SIMPLEX,
                0.7, 2
            )[0]

            cv2.putText(display_img, label_right,
                        (OUTPUT_SIZE - text_size[0] - 10, 25),
                        cv2.FONT_HERSHEY_SIMPLEX,
                        0.7, (0, 0, 255), 2)

            # Save frame
            out_dir = os.path.join(OUTPUT_DIR, best_person)
            os.makedirs(out_dir, exist_ok=True)

            saved_count[best_person] += 1

            img_path = os.path.join(
                out_dir,
                f"{best_person}_frame{frame_id}_{saved_count[best_person]}.jpg"
            )

            cv2.imwrite(img_path, display_img)

            # Write to video
            writer = get_writer(best_person)
            writer.write(display_img)

            print(f"[SAVED] {best_person} | sim={best_score:.3f} | emotion={emotion}")


#
# CLEANUP
#

cap.release()

for writer in video_writers.values():
    writer.release()

print("\nFINAL RESULTS")
for p in MEMBERS:
    print(p, ":", saved_count[p])

print("\n[INFO] Member videos saved inside:", VIDEO_OUT_DIR)


#
# DOWNLOAD FINAL OUTPUT FILES (Google Colab)
#

print("\n[INFO] Creating ZIP files for download...\n")

import shutil
from google.colab import files

shutil.make_archive("face_frames_output", 'zip', OUTPUT_DIR)
shutil.make_archive("member_videos_output", 'zip', VIDEO_OUT_DIR)

print("[INFO] Downloading ZIP files...\n")

files.download("face_frames_output.zip")
files.download("member_videos_output.zip")
```
